Assignment:

1. Pick a commonly-found probability distribution, e.g., uniform distribution or normal
distribution.
2. Use a random number generator to produce N random numbers that follow the chosen
probability distribution, where N takes the values from 10 to 200 with a step of 10.
3. Calculate the mean and standard deviation of each set of the random numbers, and make
plots to show how the mean and standard deviation vary with N.
4. Make separate histograms for N = 20, 50, 100 and 200. Compare the histogram with the
probability distribution

```python
### Imports
import numpy as np
import scipy.stats as stat
from scipy.stats import norm
import matplotlib.pyplot as plt

### I choose the normal distribution 
### Calculate answer to question 2
meanr = []
sdvr = []
Nspace = np.linspace(10, 200, 20)
for i in Nspace: 
    data = np.random.randn(int(i))
    mean = np.mean(data)
    sdv = np.sqrt(np.var(data, ddof=1))
    meanr.append(mean)
    sdvr.append(sdv)


```

```python
### Plot for question 3
fig = plt.figure(figsize = (15,9))
plt.plot(Nspace, meanr, label='Mean', marker = 'o')
plt.plot(Nspace, sdvr, label='Standard Deviation', marker = 'x')
plt.rcParams.update({'font.size': 22})
plt.xlabel('Number of Random Samples in Set')
plt.ylabel('Value of Mean and SD')
plt.rcParams.update({'font.size': 24})
plt.title(r'$\mu$ and $\sigma$ of Random Samples of the Normal Distribution')
plt.legend()
plt.savefig('MSDRand.pdf')
plt.show()
```

```python
### Question 4, calculate and plot
x = np.linspace(-4, 4, 1000)
p = norm.pdf(x, 0, 1)
histlist = [20, 50, 100, 200]
for i in histlist:
    data = np.random.randn(int(i))
    fig = plt.figure(figsize = (15,9))
    plt.hist(data, 15, density = 'true')
    plt.plot(x, p)
    plt.xlim(-4, 4)
    plt.rcParams.update({'font.size': 22})
    plt.xlabel('Generated Value')
    plt.ylabel('Normalized Counts')
    plt.rcParams.update({'font.size': 28})
    plt.title('Data vs Normal Distribution for N = %.2f' % i)
    plt.savefig('DVN_%.2f.pdf' %i)
    plt.show()
    
    
```
